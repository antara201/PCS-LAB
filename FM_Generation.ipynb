```python
import numpy as np 
import matplotlib.pyplot as plot 
A_b = float(input("Enter Baseband Amplitude: "))  
f_b = float(input("Enter Baseband Frequency: "))  
A_c = float(input("Enter Carrier Amplitude: "))  
f_c = float(input("Enter Carrier Frequency: "))  
kf = float(input("Enter modulation sensitivity (kf): "))  
t = np.linspace(0, 100, 1000) 
mu = (kf*A_b)/f_b
baseband = A_b*np.cos(2*np.pi*f_b*t) 
carrier = A_c*np.cos(2*np.pi*f_c*t) 
frequency_modulation = A_c*np.cos((2*np.pi*f_c*t) +  (mu*np.sin(2*np.pi*f_b*t))) 
plot.subplot(3,1,1) 
plot.title("Baseband signal (m(t))" ) 
plot.plot(baseband,"red") 
plot.xlabel("Time") 
plot.ylabel("Amplitude") 
plot.subplot(3,1,2) 
plot.title("Carrier signal (c(t))" ) 
plot.plot(carrier, "purple") 
plot.xlabel("Time") 
plot.ylabel("Amplitude") 
plot.subplot(3,1,3) 
plot.title("FM Modulated (FM(t))" ) 
plot.plot(frequency_modulation, color="yellow")
plot.xlabel("Time") 
plot.ylabel("Amplitude")
plot.subplots_adjust(hspace=1) 
plot.rc('font', size=15) 
fig = plot.gcf() 
fig.set_size_inches(16, 11) 
fig.suptitle("Frequency Modulation") 
```
