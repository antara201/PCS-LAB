```python
import numpy as np 
import matplotlib.pyplot as plot 
A_b = float(input("Enter Baseband Amplitude: "))  
f_b = float(input("Enter Baseband Frequency: "))  
A_c = float(input("Enter Carrier Amplitude: "))  
f_c = float(input("Enter Carrier Frequency: "))  
mu = float(input("Enter Modulation Index: "))
t = np.linspace(0, 1, 10000) 
baseband = A_b*np.cos(2*np.pi*f_b*t) 
carrier = A_c*np.cos(2*np.pi*f_c*t) 
a_m = baseband*carrier
am_wave_final= A_c * (1 + (mu*np.cos(2*np.pi*f_b*t))) *  np.cos(2*np.pi*f_c*t) 
plot.subplot(4,1,1) 
plot.title("Baseband signal (m(t))") 
plot.xlabel("Time") 
plot.plot(baseband,'g') 
plot.subplot(4,1,2) 
plot.title("Carrier signal (c(t))") 
plot.xlabel("Time") 
plot.plot(carrier, 'r') 
plot.subplot(4,1,3) 
plot.title("AM signal (DSB-SC(t))") 
plot.plot(a_m, color="yellow") 
plot.xlabel("Time") 
plot.ylabel("Amplitude") 
plot.subplot(4,1,4) 
plot.title("AM signal (AM(t))") 
plot.plot(am_wave_final, color="blue") 
plot.xlabel("Time") 
plot.ylabel("Amplitude")
plot.subplots_adjust(hspace=1) 
plot.rc('font', size=15) 
fig = plot.gcf() 
fig.suptitle("Amplitude Modulation") 
fig.set_size_inches(16, 11) 
```
